```python
import psycopg2
import psycopg2.extras

host_name = "localhost"
database = "coursebase"
username = "postgres"
port = "5432"
pwd = "87214593"

conn = None
cur = None
```

```python
try:
    conn = psycopg2.connect(
        host=host_name,
        database=database,
        user=username,
        password=pwd,
        port=port)
    print('Connected!!!')

    # Use the 'cursor' method with 'cursor_factory' parameter
    cur = conn.cursor(cursor_factory=psycopg2.extras.DictCursor)

    cur.execute('SELECT * FROM person')
    records = cur.fetchall()

    for record in records:
        print(record)

except Exception as error:
    print('Error message:', error)

finally:
    # Committing any changes and closing the connection
    if conn is not None:
        conn.commit()
        cur.close()
        conn.close()
```
